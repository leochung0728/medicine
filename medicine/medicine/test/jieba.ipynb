```python
import os
import re
import sys
sys.path.insert(1, '../..')
import pandas as pd
from medicine import config
```

```python
file_path = os.path.join(config.Path['data_dir'], '本草綱目_Ver1.csv')
data = pd.read_csv(file_path, index_col=[0], engine='python', encoding='utf-8-sig')
data.head(5)
```

```python
import jieba
jieba.set_dictionary('./jieba_data/dict.txt.big')
jieba.load_userdict('./jieba_data/userdict.txt')
```

```python
cut_result = []

for index, row in data.iterrows():
    cut = []
    word_list = list(jieba.cut(row['內文'], cut_all=False))
    
    for word in word_list:
        if not re.search(r'[\s：。，、《》∶；？！（）【】〔〕]', word):
            cut.append(word)
    cut_result.append(' '.join(cut))
```
