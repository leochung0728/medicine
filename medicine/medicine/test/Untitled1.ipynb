```python
import pandas as pd

df1 = pd.DataFrame(columns=["藥名", "釋名", "氣味", "主治", "附方"])

rows = []
with open('./data/本草綱目.txt', 'r') as file:
    for line in file.readlines()[2406:]:
        if line.split():
            for x in line.replace(" ","").split("\n"):
                if x:
                    rows.append(x)
        else:
            if len(rows) > 2:
                n = len(rows)
                while len(rows):
                    x = rows.pop(0)
                    if "【釋名】" in x:
                        #print("【釋名】", x.split("【釋名】")[1])
                        for y in x.split("【釋名】"):
                            if y:
                                df1.loc[df1.shape[0]]["釋名"] += y
                    elif "【氣味】" in x:
                        #print("【氣味】", x.split("【氣味】")[1])
                        for y in x.split("【氣味】"):
                            if y:
                                df1.loc[df1.shape[0]]["氣味"] += y
                    elif "【主治】" in x:
                        #print("【主治】", x.split("【主治】")[0], "," , x.split("【主治】")[1])
                        for y in x.split("【主治】"):
                            if y:
                                df1.loc[df1.shape[0]]["主治"] += y + ":"
                        df1.loc[df1.shape[0]]["主治"] += ","
                    elif "【附方】" in x:
                        for y in x.split("【附方】"):
                            if y:
                                #print("【附方】", y)
                                df1.loc[df1.shape[0]]["附方"] += "," + y
                        if len(rows):
                            rows[0] += "【附方】"
                    if n - len(rows) == 1:
                        #print("【藥名】", x)
                        df1.loc[df1.shape[0] + 1] = {"藥名" : "", "釋名" : "", "氣味" : "", "主治" : "", "附方" : ""}
                        df1.loc[df1.shape[0]]["藥名"] = x
            rows = []
    if len(rows) > 2:
        n = len(rows)
        while len(rows):
            x = rows.pop(0)
            if "【釋名】" in x:
                #print("【釋名】", x.split("【釋名】")[1])
                for y in x.split("【釋名】"):
                    if y:
                        df1.loc[df1.shape[0]]["釋名"] += y
            elif "【氣味】" in x:
                #print("【氣味】", x.split("【氣味】")[1])
                for y in x.split("【氣味】"):
                    if y:
                        df1.loc[df1.shape[0]]["氣味"] += y
            elif "【主治】" in x:
                #print("【主治】", x.split("【主治】")[0], "," , x.split("【主治】")[1])
                for y in x.split("【主治】"):
                    if y:
                        df1.loc[df1.shape[0]]["主治"] += y + ":"
                df1.loc[df1.shape[0]]["主治"] += ","
            elif "【附方】" in x:
                for y in x.split("【附方】"):
                    if y:
                        #print("【附方】", y)
                        df1.loc[df1.shape[0]]["附方"] += "," + y
                    if len(rows):
                        rows[0] += "【附方】"
            if n - len(rows) == 1:
                #print("【藥名】", x)
                df1.loc[df1.shape[0] + 1] = {"藥名" : "", "釋名" : "", "氣味" : "", "主治" : "", "附方" : ""}
                df1.loc[df1.shape[0]]["藥名"] = x
#df1.loc[df1.shape[0] + 1] = {"藥名" : 1, "釋名" : [1, 2, 3]}

#df1.loc[1]['col2'].extend((4,5))

# df1.loc[df1.shape[0] + 1] =  {"藥名" : '1', "釋名" : [1, 2, 3], "氣味" : "", "主治" : "", "附方" : ""}
# df1.loc[df1.shape[0]]["氣味"] = "1"
# df"1.loc[df1.shape[0]]1
df2 = df1.drop(df1[df1["釋名"] == ""][df1["氣味"] == ""][df1["主治"] == ""][df1["附方"] == ""].index)
```
